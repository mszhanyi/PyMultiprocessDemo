Depending on the platform, multiprocessing supports 2 ways to start a standalone process, spawn and fork. <br>
There are major differences. <br>

Action|fork |spawn
:-----|:---:|:---:
Import module at start of each child process|NO |YES
Variables have same id as in parent process |YES|NO 
Child process gets variables defined in name == main block |YES|NO 

```python
# see the script
with open('simplemp.py', 'r') as text:
    textfile = text.read()
    print(textfile)
```

```python
# run it
from subprocess import Popen
Popen(["python", "simplemp.py"])
```

You can see the runtime Error in Command Window
